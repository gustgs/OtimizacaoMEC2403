```python
import numpy as np
import matplotlib.pyplot as plt
import linear_search_methods as lsm

def f(P):
    return P[0]**2 - 3*P[0]*P[1] + 4*(P[1]**2) + P[0] - P[1]

def mcCormick(P):
    return np.sin(P[0] + P[1]) + (P[0] - P[1])**2 - 1.5*P[0] + 2.5*P[1]

def himmelblau(P):
    return (P[0]**2 + P[1] - 11)**2 + (P[0] + P[1]**2 - 7)**2

def dirUnit(direcao):
    return direcao/np.linalg.norm(direcao)
```

```python
P1 = np.array([1, 2])
dir = np.array([-1, -2])
q2_a_pss_ct = lsm.passo_cte(dirUnit(dir), P1, f)
intervalo = lsm.passo_cte(dirUnit(dir), P1, f)[1]
q2_a_bssc = lsm.bissecao(intervalo.copy(), dirUnit(dir), P1, f) 
q2_a_sc_ar = lsm.secao_aurea(intervalo.copy(), dirUnit(dir), P1, f)
Pmin = q2_a_sc_ar[0]

x1 = np.linspace(-2, 2, 100)
x2 = np.linspace(-3, 3, 100)
P = np.array([x1, x2])
X1, X2 = np.meshgrid(x1, x2)
x3 = f([X1, X2])
niveis = plt.contour(X1, X2, x3)
plt.clabel(niveis, inline=1, fontsize=10)
plt.annotate(f'({P1[0]}, {P1[1]})', xy=Pmin, xytext=P1, arrowprops=dict(width=0.6, color='red', headwidth=10, headlength=10), horizontalalignment='center', fontsize='10', color='red')
plt.annotate(f'  ({round(Pmin[0], 7)}, {round(Pmin[1], 7)})', xy=Pmin, xytext=Pmin, color='red', va="top", ha="left")
plt.show()
```

```python
P1 = np.array([-2, 3])
dir= np.array([1.453, -4.547])
q2_b_pss_ct = lsm.passo_cte(dirUnit(dir), P1, mcCormick)
intervalo = lsm.passo_cte(dirUnit(dir), P1, mcCormick)[1]
q2_b_bssc = lsm.bissecao(intervalo.copy(), dirUnit(dir), P1, mcCormick) 
q2_b_sc_ar = lsm.secao_aurea(intervalo.copy(), dirUnit(dir), P1, mcCormick)
Pmin = q2_b_sc_ar[0]

x1 = np.linspace(-2.5, 2.5, 100)
x2 = np.linspace(-4, 4, 100)
P = np.array([x1, x2])
X1, X2 = np.meshgrid(x1, x2)
x3 = mcCormick([X1, X2])
niveis = plt.contour(X1, X2, x3)
plt.clabel(niveis, inline=1, fontsize=10)
plt.annotate('', xy=Pmin, xytext=P1,
                arrowprops=dict(width=0.6, color='red', headwidth=10, headlength=10), fontsize='10', color='red')

plt.annotate(f'({P1[0]}, {P1[1]})', xy=P1, xytext=(0,10), textcoords='offset points', color='red', va="top", ha="left")
plt.annotate(f'({round(Pmin[0], 7)}, {round(Pmin[1], 7)})', xy=Pmin, xytext=(-50,-5), textcoords='offset points', color='red', va="top", ha="left")
plt.show()
```

```python
P1 = np.array([0, 5])
dir = np.array([3, 1.5])
q2_c_pss_ct = lsm.passo_cte(dirUnit(dir), P1, himmelblau)
intervalo = lsm.passo_cte(dirUnit(dir), P1, himmelblau)[1]
q2_c_bssc = lsm.bissecao(intervalo.copy(), dirUnit(dir), P1, himmelblau) 
q2_c_sc_ar = lsm.secao_aurea(intervalo.copy(), dirUnit(dir), P1, himmelblau)
Pmin = q2_c_sc_ar[0]

x1 = np.linspace(-0.00001, 0.00001, 200)
x2 = np.linspace(4.99999, 5.00001, 200)
P = np.array([x1, x2])
X1, X2 = np.meshgrid(x1, x2)
x3 = himmelblau([X1, X2])
niveis = plt.contour(X1, X2, x3)
plt.clabel(niveis, inline=1, fontsize=10)
plt.annotate('', xy=Pmin, xytext=P1,
                arrowprops=dict(width=0.6, color='red', headwidth=10, headlength=10), fontsize='10', color='red')

plt.annotate(f'({P1[0]}, {P1[1]})', xy=P1, xytext=(-30,15), textcoords='offset points', color='red', va="top", ha="left")
plt.annotate(f'({round(Pmin[0], 7)}, {round(Pmin[1], 7)})', xy=Pmin, xytext=(-30,20), textcoords='offset points', color='red')
plt.show()
```
