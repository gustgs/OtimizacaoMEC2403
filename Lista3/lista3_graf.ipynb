```python
import numpy as np
import matplotlib.pyplot as plt
```

### Quest√£o 2

min $f(x) = (x_1 - 1)^2 + (x_2 - 2)^2$

s.t.

$-x_1^2 + x_2 \leqslant 0$

$x_1 + x_2 - 2 = 0$

```python
def f(x):
    return (x[0] - 1)**2 + (x[1] - 2)**2

def h(x1):
    return 2-  x1

def c(x1):
    return x1**2


x1 = np.linspace(-4, 7, 1000)
x2 = np.linspace(-4, 7, 1000)
X1, X2 = np.meshgrid(x1, x2)
x3 = f([X1, X2])

y1 = np.linspace(-4, 7, 1000)
y2 = h(y1)

z1 = np.linspace(-4, 7, 1000)
z2 = c(z1)

niveis = plt.contour(X1, X2, x3,[1,4,9,16,25], colors='black')
plt.clabel(niveis, inline=1, fontsize=10)
plt.plot(y1,y2)
plt.plot(z1,z2)
plt.xlabel('$x_1$', fontsize='16')
plt.ylabel('$x_2$', fontsize='16')
plt.xlim([-4, 7])
plt.ylim([-4, 7])
plt.grid(linestyle='--')
ax = plt.gca()
ax.set_aspect(1)
plt.show()
```
